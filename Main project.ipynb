Digital literacy dataset

The Digital Literacy Dataset provides insights into individuals access to digital devices, internet usage, and basic technological skills. It helps analyze digital awareness levels and identify gaps across different demographic groups.

```python
categorical_cols = ['Gender','Education_Level','Employment_Status','Household_Income','Location_Type','Engagement_Level','Employment_Impact']
df[categorical_cols] = df[categorical_cols].astype('category')
```

Categorical variables such as gender, education level, and employment status were converted from object type to category to improve memory efficiency and facilitate better analysis. Numerical variables were already in suitable integer and float formats and required no correction.

```python
df['Computer_Knowledge_Improvement'] = (df['Post_Training_Basic_Computer_Knowledge_Score'] - df['Basic_Computer_Knowledge_Score'])
```

Training impact before vs after.
This derived column measures the improvement in basic computer knowledge after training.

```python
df['Internet_Usage_Improvement']=(df['Post_Training_Internet_Usage_Score']-df['Internet_Usage_Score'])
```

This column represents the increase in internet usage skills after training.

```python
df['Mobile_Literacy_Improvement']=(df['Post_Training_Mobile_Literacy_Score']-df['Mobile_Literacy_Score'])
```

This column captures the improvement in mobile literacy following the training program.

 Engagement & Performance Analysis

Age distribution:

```python
plt.figure()
plt.hist(df['Age'], bins=10,edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.tight_layout()
plt.show()
```

This histogram shows the distribution of participants across different age groups.
It helps identify which age range is most represented in the dataset.
The distribution provides insight into the demographic profile of learners.

```python
plt.figure()
plt.hist(df['Overall_Literacy_Score'], bins=10,edgecolor='black')
plt.xlabel('Overall Literacy Score')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Literacy Score')
plt.show()
```

```python
plt.figure()
df['Engagement_Level'].value_counts().plot(kind='bar')
plt.xlabel('Engagement Level')
plt.ylabel('Count')
plt.title('Engagement Level Distribution')
plt.show()
```

This bar chart shows the number of participants in each engagement category.
It helps assess overall involvement in the digital literacy program.
Higher engagement levels indicate stronger learner participation.

```python
plt.figure()
sns.boxplot(x='Education_Level', y='Overall_Literacy_Score', data=df)
plt.xticks(rotation=45)
plt.title('Overall Literacy Score by Education Level')
plt.show()
```

This box plot compares literacy scores across different education levels.
It highlights variations in median scores and score dispersion.
The visualization shows how educational background influences digital literacy.

```python
plt.figure()
sns.barplot(x='Gender', y='Quiz_Performance', data=df)
plt.title('Quiz Performance by Gender')
plt.show()
```

This chart compares average quiz performance across genders.
It helps identify performance differences, if any, between groups.
The visualization supports inclusive evaluation of learning outcomes.

```python
plt.figure()
sns.boxplot(x='Engagement_Level', y='Quiz_Performance', data=df)
plt.title('Quiz Performance by Engagement Level')
plt.show()
```

This box plot shows how quiz performance varies with engagement level.
It demonstrates the relationship between learner engagement and assessment results.
Higher engagement levels generally correspond to better performance.

```python
plt.figure()
plt.scatter(df['Age'], df['Overall_Literacy_Score'])
plt.xlabel('Age')
plt.ylabel('Overall Literacy Score')
plt.title('Age vs Overall Literacy Score')
plt.show()
```

This scatter plot visualizes the relationship between age and literacy score.
It helps detect trends, clusters, or age-related performance patterns.
The plot also reveals potential outliers in the data.

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr)
plt.title('Correlation Heatmap')
plt.show()
```

This heatmap shows correlations between numerical variables in the dataset.
Strong correlations highlight important relationships between features.
It is useful for understanding dependencies and supporting feature selection.

```python
plt.figure()
sns.scatterplot(
    x='Modules_Completed',
    y='Overall_Literacy_Score',
    hue='Engagement_Level',
    data=df
)

plt.title('Modules Completed vs Overall Literacy Score')
plt.show()
```

This visualization shows how module completion affects literacy outcomes.
It highlights the combined impact of engagement and learning progress.
A positive trend suggests higher completion leads to better literacy scores.

```python
improvements = df[
    [
        'Computer_Knowledge_Improvement',
        'Internet_Usage_Improvement',
        'Mobile_Literacy_Improvement'
    ]
].mean()

improvements.plot(kind='bar')
plt.ylabel('Improvement Score')
plt.title('Average Skill Improvement After Training')
plt.show()
```

This bar chart compares average improvement across digital skill areas.
It shows which skills benefited most from the training program.
The visualization confirms the effectiveness of the intervention.

```python
pre_post = [
    
        'Basic_Computer_Knowledge_Score',
        'Post_Training_Basic_Computer_Knowledge_Score',
        'Internet_Usage_Score',
        'Post_Training_Internet_Usage_Score',
        'Mobile_Literacy_Score',
        'Post_Training_Mobile_Literacy_Score'
    
]
mean_scores= df[pre_post].mean()
plt.figure(figsize=(19,6))
mean_scores.plot(kind='bar')
plt.xticks(rotation=45)
plt.ylabel('Score')
plt.title('Pre-Training vs Post-Training Average Scores')
plt.show()
```

This chart compares average skill scores before and after training.
It clearly visualizes learning gains achieved through the program.
Higher post-training scores indicate successful skill development.

```python
plt.figure()
plt.scatter(df['Session_Count'], df['Quiz_Performance'])
plt.xlabel('Session Count')
plt.ylabel('Quiz Performance')
plt.title('Session Count vs Quiz Performance')
plt.show()
```
